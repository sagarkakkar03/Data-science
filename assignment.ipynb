```python
df['Date'] = pd.to_datetime(df['Date'], format='%d-%b-%Y')
```

```python
df['week'] = df['Date'].dt.isocalendar().week.astype(int)
df['month'] = df['Date'].dt.month.astype(int)
```

## Weekly Report

```python
df.groupby('week').agg(
    total_sales = ('Sales', 'sum'),
    total_units = ('Unit', 'sum'),
    avg_sales = ('Sales', 'mean'),
    avg_units = ('Unit', 'mean')
).reset_index()
```

Monthly Report

```python
df.groupby('month').agg(
    total_sales = ('Sales', 'sum'),
    total_units = ('Unit', 'sum'),
    avg_sales = ('Sales', 'mean'),
    avg_units = ('Unit', 'mean')
).reset_index()
```

# Quaterly report

```python
df['Quarter'] = df['Date'].dt.to_period('Q')
```

```python
df.groupby('Quarter').agg(
    total_sales = ('Sales', 'sum'),
    total_units = ('Unit', 'sum'),
    avg_sales = ('Sales', 'mean'),
    avg_units = ('Unit', 'mean')
).reset_index()
```

```python
df2 = df.groupby(['State', 'Group'], as_index=False)['Sales'].sum()
fig = px.bar(df2, x="State", y="Sales", color="Group")
fig.update_layout(title="Sales by State and Group")
fig.show()
```
