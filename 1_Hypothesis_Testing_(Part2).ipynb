## Test for feature elimination

```python
from scipy.stats import shapiro
SL = 0.05
stat, p = shapiro(df['R&D Spend'])
if p < SL:
  print('Not normally distributed')
else:
  print('Normally distributed')
```

```python
stat, p = shapiro(df['Administration'])
if p < SL:
  print('Not normally distributed')
else:
  print('Normally distributed')
```

```python
from scipy.stats import ttest_ind

stat, p_value = ttest_ind(df['R&D Spend'], df['Administration'])
if p_value < SL:
  print('Features are the different')
else:
  print('Features are same')
```

```python
from scipy.stats import wilcoxon
stat, p_value = wilcoxon(df['R&D Spend'], df['Administration'])
if p_value < SL:
  print('Features are the different')
else:
  print('Features are same')
```

## Chi square test

```python
from scipy.stats import chi2_contingency
ct = pd.crosstab(df['Profit'], df['State'])
ct
```

```python
_, pvalue, _, _ = chi2_contingency(ct)
if pvalue < SL:
  print('Features are the same')
else:
  print('No similarity remove the state column')
```
