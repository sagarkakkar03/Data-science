```python
import pandas as pd
import numpy as np
```

```python
data = pd.read_csv('preprocessExample.csv')
data
```

```python
data['Country'].fillna(data['Country'].mode()[0], inplace = True)
data
```

```python
data['Age'].fillna(data['Age'].median(), inplace = True)
data['Salary'].fillna(data['Salary'].mean(), inplace = True)
data
```

```python
pd.get_dummies(data['Country'], dtype=int)
```

```python
finalDataset = pd.concat([pd.get_dummies(data['Country'], dtype=int), data.iloc[:, [1,2,3]]], axis = 1)
finalDataset
```

```python
freq = data['Country'].value_counts(normalize = True)
```

```python
data['freq'] = data['Country'].map(freq)
data
```
