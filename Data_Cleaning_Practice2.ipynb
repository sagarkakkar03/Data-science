```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv('inappropriate_data_assignment_dataset_50_records.csv')
df.head()
```

```python
df.info()
```

```python
df.drop_duplicates(inplace=True, ignore_index=True)
```

```python
df[df['Customer_ID'] <= 0]
```

```python
df[(df['Age'] <= 0) | (df['Age'] > 100)]
```

```python
df['Age'].loc[(df['Age'] <= 0) | (df['Age'] > 100)] = np.nan
```

```python
df[df['Monthly_Income'] < 0]
```

```python
df['Monthly_Income'].loc[df['Monthly_Income'] < 0] = np.nan
```

```python
df['Credit_Score'].loc[df['Credit_Score'] < 0] = np.nan
```

```python
df['Gender'].unique()
```

```python
df['Gender'].replace(['FEMALE', 'Femlae', 'male', 'female',
       'femAle', 'Male', ' MALE '], ['F', 'F', 'M', 'F', 'F', 'M', 'M'], inplace=True)
```

```python
df.head()
```

```python
print(df['Education_Level'].unique())
print(df['City'].unique())
print(df['Loan_Status'].unique())
```

```python
df[(df['City_dup'] != df['City'])]
```

```python
df.drop(columns = ['City_dup'], inplace=True)
```

```python
df['Loan_Status'].replace('Rejectd', 'rejected', inplace=True)
df['Loan_Status'] = df['Loan_Status'].apply(str.lower)
df['City'].replace('Bengaluru', 'Bangalore', inplace=True)
df['City'] = df['City'].apply(str.lower)
df['Education_Level'] = df['Education_Level'].apply(str.lower)
df['Education_Level'].replace(['undergraduate', 'post graduate', 'gradute', 'under graduate'], ['ug', 'pg', 'graduate', 'ug'], inplace=True)
```
