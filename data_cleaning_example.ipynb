```python
import numpy as np
import pandas as pd
```

```python
df = pd.read_csv('datasetExample.csv')
```

```python
df
```

```python
df.info()
```

```python
df.drop_duplicates(inplace=True, ignore_index=True)
```

```python
df
```

```python
df[df['Age_Group'] != df['Age_Group.1']]
```

```python
df.drop(columns = ['Age_Group.1'], inplace=True)
```

```python
df
```

```python
df['Bill'].loc[df['Bill'] < 0] = np.nan
```

```python
df
```

```python
df['Rating(1-5)'].loc[(df['Rating(1-5)'] >= 6) | (df['Rating(1-5)'] <= 0)] = np.nan
```

```python
df
```

```python
df['EstimatedSalary'].loc[df['EstimatedSalary'] <= 0] = np.nan
```

```python
df
```

```python
df['NoOfPax'].loc[df['NoOfPax'] < 0] = np.nan
```

```python
df
```

```python
df['Hotel'].unique()
```

```python
df['Hotel'].replace('Ibys', 'ibis', inplace=True)
```

```python
df
```

```python
df['Hotel'] = df['Hotel'].apply(str.lower)
```

```python
df['FoodPreference'].replace('Vegetarian', 'veg', inplace=True)
df['FoodPreference'] = df['FoodPreference'].apply(str.lower)
df
```
